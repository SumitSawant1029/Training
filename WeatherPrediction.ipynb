***Normalization*** We Will choose Standardization as the data consist of ouliers

```python
df.head(1)
```

```python
# Standardize the 'Temp' column in-place
df["Rel Hum_%"] = (df["Rel Hum_%"] - df["Rel Hum_%"].mean()) / df["Rel Hum_%"].std()
# Standardize the 'Temp' column in-place
df["Wind Speed_km/h"] = (df["Wind Speed_km/h"] - df["Wind Speed_km/h"].mean()) / df["Wind Speed_km/h"].std()
df["Dew Point Temp_C"] = (df["Dew Point Temp_C"] - df["Dew Point Temp_C"].mean()) / df["Dew Point Temp_C"].std()

df["Press_kPa"] = (df["Press_kPa"] - df["Press_kPa"].mean()) / df["Press_kPa"].std()
df["Temp_C"] = (df["Temp_C"] - df["Temp_C"].mean()) / df["Temp_C"].std()
```

```python


# Function to group weather categories
def simplify_weather(label):
    label = label.lower()
    if "fog" in label:
        return "Fog"
    elif "snow" in label:
        return "Snow"
    elif "rain" in label or "drizzle" in label:
        return "Rain"
    elif "clear" in label or "cloudy" in label or "haze" in label:
        return "Clear"
    else:
        return "Other"

# Apply grouping
df["Weather"] = df["Weather"].apply(simplify_weather)
```

***Implementing Model***

**Random Forest**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
X = df.drop(columns=["Weather","Date/Time"])
y=df["Weather"]
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X
```

```python
y
```

```python
model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
