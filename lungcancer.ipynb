```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x= df.drop(columns='LUNG_CANCER')
y = df['LUNG_CANCER']
X_train, X_test ,Y_train ,Y_test = train_test_split(x,y,test_size=0.5,random_state=102)
```

```python
X_train.head()
```

DECISION TREE IMPLEMENTATION

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score ,classification_report
import matplotlib.pyplot as plt

# Train the model
model = DecisionTreeClassifier(max_depth=4,random_state=42)
model.fit(X_train, Y_train)

# Make predictions
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(Y_test, y_pred))
print("Accuracy:", classification_report(Y_test, y_pred))
# Visualize the decision tree
plt.figure(figsize=(20, 10))
plot_tree(
    model,
    filled=True,
    feature_names=X_train.columns,                   
    class_names=[str(cls) for cls in model.classes_] 
)
plt.title("Decision Tree Visualization")
plt.show()
```

RANDOM FOREST IMPLEMENTATION

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100,random_state=0)
rf_model.fit(X_train,Y_train)

rf_pred = rf_model.predict(X_test)
print("Accuracy:", accuracy_score(Y_test, rf_pred))
```

```python
# Fill NaNs in 'PEER_PRESSURE' with the mean value of that column
X_train['PEER_PRESSURE'].fillna(1, inplace=True)
```

K-Mean Clustering (KMeans)  --> UnSupervised Learning Algorithm --> Clustering 

```python
# from sklearn.cluster import KMeans

# modelkmean = KMeans()
# modelkmean.fit(X_train)
# clusters = modelkmean.predict(X_test)

# plt.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=clusters, cmap='viridis')
# plt.xlabel('Feature 1')
# plt.ylabel('Feature 2')
# plt.title('KMeans Clustering')
# plt.show()


from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Fit KMeans with a reasonable number of clusters
modelkmean = KMeans(n_clusters=3, random_state=42)
modelkmean.fit(X_train_scaled)
clusters = modelkmean.predict(X_test_scaled)

# Plot
plt.scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Feature 1 (scaled)')
plt.ylabel('Feature 2 (scaled)')
plt.title('KMeans Clustering (Scaled Features)')
plt.show()
```

K Nearest Neighbour (KNN)   --> Supervised Learning Algorithm  --> Classification

```python
from sklearn.neighbors import KNeighborsClassifier

modelKNN = KNeighborsClassifier(n_neighbors=3)
modelKNN.fit(X_train,Y_train)
```

```python
accuracy = modelKNN.score(X_test, Y_test)
print(f"Accuracy on test set: {accuracy:.2f}")
```

```python
y_pred = modelKNN.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

print(classification_report(Y_test, y_pred))

cm = confusion_matrix(Y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
